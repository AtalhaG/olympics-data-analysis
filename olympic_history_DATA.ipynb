```python
def medal_point(medal):
    if medal=="Gold":
        return 3
    elif medal=="Silver":
        return 2
    elif medal=="Bronze":
        return 1
    else:
        return 0
```

```python
data["MedalPoint"]=data["Medal"].apply(medal_point)
```

```python
data
```

```python

data[data["MedalPoint"]>2]
```

```python
data[(data["Team"]=="Turkey")&(data["MedalPoint"]>0)]
```

```python
turkey_filtered=data[(data["Team"]=="Turkey")&(data["MedalPoint"]>0)&(data["Year"]>2000)]
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height",y="Weight",hue="Sex",data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight",hue="Medal",data=data)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height",y="Weight",hue="Sex",style="Medal",size="Age",data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight",hue="Sex",data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height",hue="Sex",data=data)
plt.ylabel("Frequency")
plt.xlabel("Height Distribution")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height",hue="Sex",data=data,kind="kde")
plt.ylabel("Frequency")
plt.xlabel("Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal",y="Height",hue="Sex",data=data)
plt.show()
```
